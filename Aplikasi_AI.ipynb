```python
# Import Library
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd 
data_jantung = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
data_jantung['DEATH_EVENT'].value_counts()
```

```python
# memisahkan data dan label 
x = data_jantung.drop(columns='DEATH_EVENT', axis=1)
y = data_jantung['DEATH_EVENT']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
scaler.fit(x)
```

```python
standarized_data = scaler.transform(x)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(standarized_data, y)
```

```python
print(standarized_data)
```

```python
x = standarized_data
y = data_jantung['DEATH_EVENT']
```

```python
print(x)
print(y)
```

```python
x_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
print(x.shape, x_train.shape, X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
# Evaluasi akurasi pada data training
x_train_prediction = classifier.predict(x_train)  # Menggunakan X_train, bukan x_train
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
```

```python
print('Akurasi Data Training Adalah = ', training_data_accuracy)
```

```python
y_test_prediction = classifier.predict(X_test)  # Prediksi berdasarkan X_test
test_data_accuracy = accuracy_score(y_test, y_test_prediction)
```

```python
print('Akurasi Data Testing Adalah = ', test_data_accuracy)
```

```python
# Data input baru 
input_data = (53,0,63,1,60,0,368000,0.8,135,1,0,22)

# Konversi ke numpy array
input_data_as_numpy_array = np.array(input_data)

# Ubah menjadi array 2D
input_data_reshape = input_data_as_numpy_array.reshape(1, -1)

# Standardisasi data input
std_data = scaler.transform(input_data_reshape)

# Prediksi
prediction = classifier.predict(std_data)
print(prediction)

# Interpretasi hasil
if prediction[0] == 1:
    print('Pasien terkena Jantung')
else:
    print('Pasien tidak terkena Jantung')
```
